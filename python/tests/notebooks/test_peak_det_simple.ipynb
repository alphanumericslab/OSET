```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from oset.generic.lp_filter.lp_filter_zero_phase import lp_filter_zero_phase
from oset.ecg.peak_detection.peak_det_simple import peak_det_simple
```

```python
mat = scipy.io.loadmat("../../../datasets/sample-data/SampleECG1.mat")["data"][0]
f = 1
fs = 1000
fc = 0.5
t = np.arange(len(mat)) / fs
mat = mat - lp_filter_zero_phase(mat, fc / fs)
peaks, peak_indexes = peak_det_simple(mat, f / fs)
```

```python
plt.figure()
plt.figure(figsize=(12, 5))
plt.plot(t, mat, "b", label="ECG")
plt.plot(t[peak_indexes], mat[peak_indexes], "ro", label="ECG Peaks (max detection)")
plt.xlabel("time (sec.)")
plt.legend()
plt.grid()
plt.show()
```

```python
# top peaks
peaks, peak_indexes = peak_det_simple(mat, f / fs, 1)
```

```python
plt.figure()
plt.figure(figsize=(12, 5))
plt.plot(t, mat, "b", label="ECG")
plt.plot(t[peak_indexes], mat[peak_indexes], "ro", label="ECG Peaks (max detection)")
plt.xlabel("time (sec.)")
plt.legend()
plt.grid()
plt.show()
```
