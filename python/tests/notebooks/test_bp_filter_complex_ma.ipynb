```python
from oset.generic import bp_filter_complex_ma
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
n_samples = 1000
t = np.linspace(0, 1, n_samples)
signal = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 20 * t)  # A signal composed of 10 Hz and 20 Hz components
x = np.tile(signal, (3, 1))  #3 channels with the same signal

# Normalize frequencies by sampling frequency
fs = 1000  # Sampling frequency
fc = 15 / fs  # Center frequency (normalized)
bw = 10 / fs  # Bandwidth (normalized)
order = 2  # Order of the filter

# Apply the bandpass filter
y = bp_filter_complex_ma(x, fc, bw, order)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x[0, :])
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(t, y[0, :])
plt.title('Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
