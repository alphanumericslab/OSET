```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from peak_detection_amp_threshold.peak_detection_amp_threshold import peak_detection_amp_threshold
```

```python
mat = scipy.io.loadmat('../../../datasets/sample-data/SampleECG1.mat')['data'][0]
f = 1
fs = 1000
t = np.arange(len(mat)) / fs
peaks=peak_detection_amp_threshold(mat, f / fs, 0.0)[0]
```

```python
plt.figure()
plt.figure(figsize=(12,5))
plt.plot(t, mat, 'b', label='ECG')
plt.plot(t, peaks * mat, 'ro', label='ECG Peaks (max detection)')
plt.xlabel('time (sec.)')
plt.legend()
plt.grid()
plt.show()
```

```python
# neglect some peaks
peaks=peak_detection_amp_threshold(mat, f / fs, 0.9)[0]
plt.figure()
plt.figure(figsize=(12,5))
plt.plot(t, mat, 'b', label='ECG')
plt.plot(t, peaks * mat, 'ro', label='ECG Peaks (max detection)')
plt.xlabel('time (sec.)')
plt.legend()
plt.grid()
plt.show()
```
