```python
import pandas as pd
import numpy as np
import os

file_name = 'ResultsComparison120s'
file_path = os.getcwd()
target_csv_path = file_path + '/../../../../' + file_name + '/'
fs = 250
reps_range = [1,2,3,4,5]

for rep in reps_range:
    print("starting the QT dataframe construction for rep =", rep)            
    QT_csv_name = 'QTRep' + str(rep)
    
    #------------------------------------
    print("    reading the clean signals")            
    csv_nameRaw = 'QTMediansRdb_' + str(rep)
    dfRaw = pd.read_csv(target_csv_path + csv_nameRaw + '.csv', header = None, index_col = None)

    #------------------------------------
    print("    reading the wavelet filtered signals")            
    csv_name30 = 'QTMediansWaveFilt30db_' + str(rep)
    df30 = pd.read_csv(target_csv_path + csv_name30 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name25 = 'QTMediansWaveFilt25db_' + str(rep)
    df25 = pd.read_csv(target_csv_path + csv_name25 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name20 = 'QTMediansWaveFilt20db_' + str(rep)
    df20 = pd.read_csv(target_csv_path + csv_name20 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name15 = 'QTMediansWaveFilt15db_' + str(rep)
    df15 = pd.read_csv(target_csv_path + csv_name15 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name10 = 'QTMediansWaveFilt10db_' + str(rep)
    df10 = pd.read_csv(target_csv_path + csv_name10 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name5 = 'QTMediansWaveFilt5db_' + str(rep)
    df5 = pd.read_csv(target_csv_path + csv_name5 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name0 = 'QTMediansWaveFilt0db_' + str(rep)
    df0 = pd.read_csv(target_csv_path + csv_name0 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_nameMin5 = 'QTMediansWaveFilt-5db_' + str(rep)
    dfMin5 = pd.read_csv(target_csv_path + csv_nameMin5 + '.csv', header = None, index_col = None)
    #------------------------------------
    print("      computing the differences data frames")            
    df30 = df30.subtract(dfRaw, axis = 1).T
    df30.columns = ['QT', 'RR']
    df30['QT'] = df30['QT']/fs*1000
    df30['RR'] = df30['RR']/1000
    df30['SNR'] = [30]*df30.shape[0]
    #------------------------------------
    df25 = df25.subtract(dfRaw, axis = 1).T
    df25.columns = ['QT', 'RR']
    df25['QT'] = df25['QT']/fs*1000
    df25['RR'] = df25['RR']/1000
    df25['SNR'] = [25]*df25.shape[0]
    #------------------------------------
    df20 = df20.subtract(dfRaw, axis = 1).T
    df20.columns = ['QT', 'RR']
    df20['QT'] = df20['QT']/fs*1000
    df20['RR'] = df20['RR']/1000
    df20['SNR'] = [20]*df20.shape[0]
    #------------------------------------
    df15 = df15.subtract(dfRaw, axis = 1).T
    df15.columns = ['QT', 'RR']
    df15['QT'] = df15['QT']/fs*1000
    df15['RR'] = df15['RR']/1000
    df15['SNR'] = [15]*df15.shape[0]
    #------------------------------------
    df10 = df10.subtract(dfRaw, axis = 1).T
    df10.columns = ['QT', 'RR']
    df10['QT'] = df10['QT']/fs*1000
    df10['RR'] = df10['RR']/1000
    df10['SNR'] = [10]*df10.shape[0]
    #------------------------------------
    df5 = df5.subtract(dfRaw, axis = 1).T
    df5.columns = ['QT', 'RR']
    df5['QT'] = df5['QT']/fs*1000
    df5['RR'] = df5['RR']/1000
    df5['SNR'] = [5]*df5.shape[0]
    #------------------------------------
    df0 = df0.subtract(dfRaw, axis = 1).T
    df0.columns = ['QT', 'RR']
    df0['QT'] = df0['QT']/fs*1000
    df0['RR'] = df0['RR']/1000
    df0['SNR'] = [0]*df0.shape[0]
    #------------------------------------
    dfMin5 = dfMin5.subtract(dfRaw, axis = 1).T
    dfMin5.columns = ['QT', 'RR']
    dfMin5['QT'] = dfMin5['QT']/fs*1000
    dfMin5['RR'] = dfMin5['RR']*1000
    dfMin5['SNR'] = [-5]*dfMin5.shape[0]
    #------------------------------------
    print("         computing the WAVE data frame")            
    WAVE = pd.concat([dfMin5, df0, df5, df10, df15, df20, df25, df30], axis = 0)
    WAVE['Method'] = ['Wavelet']*WAVE.shape[0]

    
    print("    reading the GP filtered signals")     
    csv_name30 = 'QTMediansGPFiltPost30db_' + str(rep)
    df30 = pd.read_csv(target_csv_path + csv_name30 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name25 = 'QTMediansGPFiltPost25db_' + str(rep)
    df25 = pd.read_csv(target_csv_path + csv_name25 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name20 = 'QTMediansGPFiltPost20db_' + str(rep)
    df20 = pd.read_csv(target_csv_path + csv_name20 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name15 = 'QTMediansGPFiltPost15db_' + str(rep)
    df15 = pd.read_csv(target_csv_path + csv_name15 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name10 = 'QTMediansGPFiltPost10db_' + str(rep)
    df10 = pd.read_csv(target_csv_path + csv_name10 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name5 = 'QTMediansGPFiltPost5db_' + str(rep)
    df5 = pd.read_csv(target_csv_path + csv_name5 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_name0 = 'QTMediansGPFiltPost0db_' + str(rep)
    df0 = pd.read_csv(target_csv_path + csv_name0 + '.csv', header = None, index_col = None)
    #------------------------------------
    csv_nameMin5 = 'QTMediansGPFiltPost-5db_' + str(rep)
    dfMin5 = pd.read_csv(target_csv_path + csv_nameMin5 + '.csv', header = None, index_col = None)
    #------------------------------------
    print("      computing the differences data frames")   
    df30 = df30.subtract(dfRaw, axis = 1).T
    df30.columns = ['QT', 'RR']
    df30['QT'] = df30['QT']/fs*1000
    df30['RR'] = df30['RR']/1000
    df30['SNR'] = [30]*df30.shape[0]
    #------------------------------------
    df25 = df25.subtract(dfRaw, axis = 1).T
    df25.columns = ['QT', 'RR']
    df25['QT'] = df25['QT']/fs*1000
    df25['RR'] = df25['RR']/1000
    df25['SNR'] = [25]*df25.shape[0]
    #------------------------------------
    df20 = df20.subtract(dfRaw, axis = 1).T
    df20.columns = ['QT', 'RR']
    df20['QT'] = df20['QT']/fs*1000
    df20['RR'] = df20['RR']/1000
    df20['SNR'] = [20]*df20.shape[0]
    #------------------------------------
    df15 = df15.subtract(dfRaw, axis = 1).T
    df15.columns = ['QT', 'RR']
    df15['QT'] = df15['QT']/fs*1000
    df15['RR'] = df15['RR']/1000
    df15['SNR'] = [15]*df15.shape[0]
    #------------------------------------
    df10 = df10.subtract(dfRaw, axis = 1).T
    df10.columns = ['QT', 'RR']
    df10['QT'] = df10['QT']/fs*1000
    df10['RR'] = df10['RR']/1000
    df10['SNR'] = [10]*df10.shape[0]
    #------------------------------------
    df5 = df5.subtract(dfRaw, axis = 1).T
    df5.columns = ['QT', 'RR']
    df5['QT'] = df5['QT']/fs*1000
    df5['RR'] = df5['RR']/1000
    df5['SNR'] = [5]*df5.shape[0]
    #------------------------------------
    df0 = df0.subtract(dfRaw, axis = 1).T
    df0.columns = ['QT', 'RR']
    df0['QT'] = df0['QT']/fs*1000
    df0['RR'] = df0['RR']/1000
    df0['SNR'] = [0]*df0.shape[0]
    #------------------------------------
    dfMin5 = dfMin5.subtract(dfRaw, axis = 1).T
    dfMin5.columns = ['QT', 'RR']
    dfMin5['QT'] = dfMin5['QT']/fs*1000
    dfMin5['RR'] = dfMin5['RR']*1000
    dfMin5['SNR'] = [-5]*dfMin5.shape[0]
    #------------------------------------
    print("         computing the GP data frame")            
    GP = pd.concat([dfMin5, df0, df5, df10, df15, df20, df25, df30], axis = 0)
    GP['Method'] = ['GP Filter']*WAVE.shape[0]
    print("      computing the QTDataFrame")            
    QTDataFrame = pd.concat([WAVE, GP], axis = 0)
    print("      saving the QTDataFrame as " + QT_csv_name)            
    QTDataFrame.to_csv(target_csv_path + QT_csv_name + '.csv')
```

```python
QTDataFrame1 = pd.read_csv(target_csv_path + 'QTRep' + str(1) + '.csv', index_col = 0)
QTDataFrame2 = pd.read_csv(target_csv_path + 'QTRep' + str(2) + '.csv', index_col = 0)
QTDataFrame3 = pd.read_csv(target_csv_path + 'QTRep' + str(3) + '.csv', index_col = 0)
QTDataFrame4 = pd.read_csv(target_csv_path + 'QTRep' + str(4) + '.csv', index_col = 0)
QTDataFrame5 = pd.read_csv(target_csv_path + 'QTRep' + str(5) + '.csv', index_col = 0)

QTMean = (QTDataFrame1['QT'] + QTDataFrame2['QT'] + QTDataFrame3['QT'] + QTDataFrame4['QT'] + QTDataFrame5['QT'])/5
RRMean = (QTDataFrame1['RR'] + QTDataFrame2['RR'] + QTDataFrame3['RR'] + QTDataFrame4['RR'] + QTDataFrame5['RR'])/5
```

```python
QTDataFrameMean = pd.DataFrame()
QTDataFrameMean['QT']=QTMean
QTDataFrameMean['RR']=RRMean
QTDataFrameMean['SNR']=QTDataFrame1['SNR']
QTDataFrameMean['Method']=QTDataFrame1['Method']
```
