```python
import pandas as pd
import numpy as np
import os
# from ComputeInputOutputDeltaSNRs import ComputeInputOutputDeltaSNRs
#-------------------------------------------------------------------------------------------------------
fs = 250
folder_name = '/../../../../ResultsComparison120s/'
repetition  = 1       # can be repetition   =  5,  4,  3,  2,  1
#-------------------------------------------------------------------------------------------------------
csv_name = 'CleanSignals' + str(repetition)
CleanDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
                        header = 0, index_col = 0)
#-------------------------------------------------------------------------------------------------------
# Select a noise level and a repetition from the csv files from the ResultsComparison30s folder
noise_level = -5      # can be noise_level  =  30,  25,  20,  15,  10,  5,  0,  -5

#--------------
# Load the Noisy signals csv file
csv_name = 'NoisySignals' + str(noise_level) + 'db_' + str(repetition)
NoisyDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
                      header = 0, index_col = 0)
#--------------
# Compute the noise signals 
NoiseDF = NoisyDF.subtract(CleanDF)
#-------------------------------------------------------------------------------------------------------
# Load the GP Posterior filtered signals csv file
csv_name = 'GPFiltPost' + str(noise_level) + 'db_' + str(repetition)
GPPostDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
                       header = 0, index_col = 0)
# Load the GP Prior filtered signals csv file
csv_name = 'GPFiltPrio' + str(noise_level) + 'db_' + str(repetition)
GPPrioDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
                       header = 0, index_col = 0)
#-------------------------------------------------------------------------------------------------------
# Load the Wavelet filtered signals csv file
csv_name = 'WaveFilt' + str(noise_level) + 'db_' + str(repetition)
WaveDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
                     header = None, index_col = 0) 
#-------------------------------------------------------------------------------------------------------
# # Load the Reza filtered Posterior filtered signals (No filtered matrix) csv file
# csv_name = 'RezaFiltPostBYPASS' + str(noise_level) + 'db_' + str(repetition)
# RezaPostByDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
# # Load the Reza filtered Prior filtered signals (No filtered matrix) csv file
# csv_name = 'RezaFiltPrioBYPASS' + str(noise_level) + 'db_' + str(repetition)
# RezaPrioByDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
# #-------------------------------------------------------------------------------------------------------
# # Load the Reza filtered Posterior filtered signals (MA filtered matrix) csv file
# csv_name = 'RezaFiltPostMA' + str(noise_level) + 'db_' + str(repetition)
# RezaPostMaDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
# # Load the Reza filtered Prior filtered signals (MA filtered matrix) csv file
# csv_name = 'RezaFiltPrioMA' + str(noise_level) + 'db_' + str(repetition)
# RezaPrioMaDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
#-------------------------------------------------------------------------------------------------------
# # Load the Reza filtered Posterior filtered signals (Gaussian filtered matrix) csv file
# csv_name = 'RezaFiltPostGAUSSIAN' + str(noise_level) + 'db_' + str(repetition)
# RezaPostGaDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
# # Load the Reza filtered Prior filtered signals (Gaussian filtered matrix) csv file
# csv_name = 'RezaFiltPrioGAUSSIAN' + str(noise_level) + 'db_' + str(repetition)
# RezaPrioGaDF = pd.read_csv(os.getcwd() + folder_name + csv_name + '.csv', 
#                          header = None, index_col = 0) 
# #-------------------------------------------------------------------------------------------------------
```

```python
from ComputeInputOutputDeltaSNRs import ComputeInputOutputDeltaSNRs
#------------------------------------------------------------------------------------------------
# Select any ECG from the list ("name" for channel 1, "name.1" for channel 2)
# 'sel30', 'sel47', 'sel14046', 'sele0603', 'sel16786', 'sel123', 'sel37', 'sele0166', 
# 'sele0509', 'sel811', 'sel16273', 'sel16795', 'sel50', 'sel307', 'sel44', 'sel41', 
# 'sele0409', 'sel873', 'sele0110', 'sele0104', 'sele0122', 'sele0136', 'sel221', 
# 'sele0111', 'sel16420', 'sel42', 'sele0210', 'sel232', 'sele0606', 'sele0612', 
# 'sele0116', 'sele0607', 'sel233', 'sel117', 'sel871', 'sele0211', 'sel43', 'sele0406', 
# 'sel45', 'sel16539', 'sel51', 'sel15814', 'sele0411', 'sel308', 'sel872', 'sel40', 
# 'sele0405', 'sele0604', 'sel301', 'sel306', 'sele0203', 'sel16483', 'sel16272', 'sel14172', 
# 'sel36', 'sel17453', 'sel104', 'sele0704', 'sele0170', 'sel17152', 'sel52', 'sel46', 
# 'sel31', 'sel803', 'sel853', 'sel847', 'sel33', 'sel49', 'sel820', 'sel808', 'sel16773', 
# 'sel34', 'sel310', 'sel230', 'sele0107', 'sel103', 'sele0121', 'sel213', 'sel231', 'sele0114',
# 'sele0303', 'sele0609', 'sele0129', 'sel223', 'sele0112', 'sele0106', 'sel35', 'sel16265', 
# 'sel14157', 'sel821', 'sel891', 'sel48', 'sel32', 'sele0124', 'sel102', 'sel116', 'sel100', 
# 'sel114', 'sel302', 'sele0133', 'sele0126', 'sel883', 'sel840', 'sel38', 'sel39'            
ecg_name  =  'sel100' # 'sel30' 'sele0612'
indx  =  CleanDF.columns.get_loc(ecg_name)
#--------------------------------------------------------------------------
Clean     =  CleanDF.iloc[:,indx].dropna()
#--------------------------------------------------------------------------
Noisy     =  NoisyDF.iloc[:,indx].dropna()
Noise     =  NoiseDF.iloc[:,indx].dropna()
#--------------------------------------------------------------------------
GPPrio    =  GPPrioDF.iloc[:,indx].dropna()
GPPost    =  GPPostDF.iloc[:,indx].dropna()
#---------
deltaSNRGPPrio  =  ComputeInputOutputDeltaSNRs(Clean, Noise, GPPrio)[2]
deltaSNRGPPost  =  ComputeInputOutputDeltaSNRs(Clean, Noise, GPPost)[2]
#--------------------------------------------------------------------------
Wave      =  WaveDF.iloc[:,indx].dropna()
#---------
deltaSNRWave = ComputeInputOutputDeltaSNRs(Clean, Noise, Wave)[2]
#--------------------------------------------------------------------------
# RezaPrioBy  =  RezaPrioByDF.iloc[:,indx].dropna()
# RezaPostBy  =  RezaPostByDF.iloc[:,indx].dropna()
# #---------
# deltaSNRRezaPrioBy  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPrioBy)[2]
# deltaSNRRezaPostBy  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPostBy)[2]
# #--------------------------------------------------------------------------
# RezaPrioMa  =  RezaPrioMaDF.iloc[:,indx].dropna()
# RezaPostMa  =  RezaPostMaDF.iloc[:,indx].dropna()
# #---------
# deltaSNRRezaPrioMa  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPrioMa)[2]
# deltaSNRRezaPostMa  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPostMa)[2]

#--------------------------------------------------------------------------
# RezaPrioGa  =  RezaPrioGaDF.iloc[:,indx].dropna()
# RezaPostGa  =  RezaPostGaDF.iloc[:,indx].dropna()
# #---------
# deltaSNRRezaPrioGa  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPrioGa)[2]
# deltaSNRRezaPostGa  =  ComputeInputOutputDeltaSNRs(Clean, Noise, RezaPostGa)[2]
# #--------------------------------------------------------------------------
```

```python
target_file = os.getcwd() + '/../../../../ResultsComparison120s/'
target_file_name = target_file + 'Comp_' + ecg_name + '_' + str(noise_level) + 'dB'
detail = 1000
fntsz = 32
#------------
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16, 11))
#--------------------------------------------------------------------------
ax = plt.subplot(3, 1, 1)
plt.plot(np.arange(len(Clean[0:detail]))/fs, Noisy[0:detail], lw = 2, label = 'x')
plt.plot(np.arange(len(Clean[0:detail]))/fs, GPPrio[0:detail], lw = 2, 
         label = 'GP Prior [ΔSNR=' + str(np.round(deltaSNRGPPrio,1)) + ']')
ax.set_xticks([], [])
#plt.xticks(fontsize = fntsz)
plt.yticks(fontsize = fntsz)
ax.set_yticks(np.arange(np.round(np.min(Noisy[0:detail]),1), 
                        np.round(np.max(Noisy[0:detail]),1) + 0.01, 
                        (np.round(max(Noisy[0:detail]),1)-np.round(min(Noisy[0:detail]),1))/5))
plt.legend(fontsize = fntsz, loc = "upper right", frameon=False, ncol = 2)
plt.ylabel("amplitude [mV]", fontsize = fntsz)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
#--------------------------------------------------------------------------
ax = plt.subplot(3, 1, 2)
plt.plot(np.arange(len(Clean[0:detail]))/fs, Noisy[0:detail], lw = 2)
plt.plot(np.arange(len(Clean[0:detail]))/fs, GPPost[0:detail], lw = 2,
         label = 'GP Posterior [ΔSNR=' + str(np.round(deltaSNRGPPost,1)) + ']')
ax.set_xticks([], [])
plt.xticks(fontsize = fntsz)
plt.yticks(fontsize = fntsz)    
ax.set_yticks(np.arange(np.round(min(Noisy[0:detail]),1), 
                        np.round(max(Noisy[0:detail]),1) + 0.01, 
                        (np.round(max(Noisy[0:detail]),1)-np.round(min(Noisy[0:detail]),1))/5))
plt.legend(fontsize = fntsz, loc = "upper right", frameon=False, ncol = 2)
plt.ylabel("amplitude [mV]", fontsize = fntsz)
#plt.xlabel("time [s]", fontsize = fntsz)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
#--------------------------------------------------------------------------
ax = plt.subplot(3, 1, 3)
plt.plot(np.arange(len(Clean[0:detail]))/fs, Noisy[0:detail], lw = 2)
plt.plot(np.arange(len(Clean[0:detail]))/fs, Wave[0:detail], lw = 2, 
         label = 'Wavelet [ΔSNR=' + str(np.round(deltaSNRWave,1)) + ']')
#ax.set_xticks([], [])
plt.xticks(fontsize = fntsz)
plt.yticks(fontsize = fntsz)    
ax.set_yticks(np.arange(np.round(min(Noisy[0:detail]),1), 
                        np.round(max(Noisy[0:detail]),1) + 0.01, 
                        (np.round(max(Noisy[0:detail]),1)-np.round(min(Noisy[0:detail]),1))/5))
plt.legend(fontsize = fntsz, loc = "upper right", frameon=False, ncol = 2)
plt.ylabel("amplitude [mV]", fontsize = fntsz)
plt.xlabel("time [s]", fontsize = fntsz)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(False)

#--------------------------------------------------------------------------
fig.tight_layout()
plt.savefig(target_file_name + 'dB.eps', format = 'eps', bbox_inches='tight')
plt.savefig(target_file_name + 'dB.png', format = 'png', bbox_inches='tight')
```
